## Background

- Spectral classification is crucial in astronomy for understanding physical properties of stars such as temperature, luminosity, and chemical composition.
- Traditional methods of classification are subjective and time-consuming as they rely on visual inspection by experts.
- Machine learning offers a promising alternative, leveraging models to learn patterns and predict spectral classes accurately.
- Large-scale astronomical datasets from SDSS and Gaia have enabled the development of machine learning models for spectral classification.

## Method - Data Analysis, feature engineering, model training

Notes: add the scatter of Gmag vs bp-rp

Method for data cleaning and column transformation:
- Obtained a dataset of 100,000 astronomical objects classified as stars from the Sloan Digital Sky Survey and cross-matched it with the Gaia dataset to extract additional features for classification purposes.
- Converted the absolute magnitude of the stars to relative magnitude to account for the effects of distance and extinction
- Simplified the target class of each star into their corresponding spectral groups (O, B, A, F, G, K, M)
- Calculated the difference in magnitude between the blue and red filters, which is commonly used as a measure of the color of the star, and added a new column for improved classification.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
# Use pd.read_csv to read file
path = "Data/result.csv"
raw_data = pd.read_csv(path)
```

Now let's check the info again.

## Dropping Unwanted Column

This Stackoverflow link will give you idea of how to drop a column from pandas dataframe.
https://stackoverflow.com/questions/13411544/delete-column-from-pandas-dataframe

```python
#The best way to do this in pandas is to use drop:
raw_data_dropped = raw_data[["umag","gmag","rmag","imag","zmag","subCl", "Plx", "Gmag", "BPmag", "RPmag"]]
raw_data_dropped = raw_data_dropped.dropna()
```

```python
#Copy the SpType column to a new column called TargetClass
raw_data_dropped['TargetClass'] = raw_data['subCl']
raw_data_dropped
```

create new columns for relative magnitude

```python
#Dropping rows that `Plx` = 0
raw_data_plx = raw_data_dropped[raw_data_dropped.Plx > 0]

#Reindex the dataframe
raw_data_plx = raw_data_plx.reset_index(drop=True)
raw_data_plx

```

```python
absolute = [["umag","gmag","rmag","imag","zmag", "Gmag", "BPmag", "RPmag"]]

raw_data_plx["r_umag"] = raw_data_plx["umag"] - 5* (np.log10(100/raw_data_plx["Plx"]))
raw_data_plx["r_gmag"] = raw_data_plx["gmag"] - 5* (np.log10(100/raw_data_plx["Plx"]))
raw_data_plx["r_rmag"] = raw_data_plx["rmag"] - 5* (np.log10(100/raw_data_plx["Plx"]))
raw_data_plx["r_imag"] = raw_data_plx["imag"] - 5* (np.log10(100/raw_data_plx["Plx"]))
raw_data_plx["r_zmag"] = raw_data_plx["zmag"] - 5* (np.log10(100/raw_data_plx["Plx"]))
raw_data_plx["r_Gmag"] = raw_data_plx["Gmag"] - 5* (np.log10(100/raw_data_plx["Plx"]))
raw_data_plx["r_BPmag"] = raw_data_plx["BPmag"] - 5* (np.log10(100/raw_data_plx["Plx"]))
raw_data_plx["r_RPmag"] = raw_data_plx["RPmag"] - 5* (np.log10(100/raw_data_plx["Plx"]))

raw_data_plx
data_clean_target = raw_data_plx
```

Cleaning target data

```python
for i in range(len(data_clean_target['TargetClass'])):
     if "O" in data_clean_target.loc[i,'TargetClass']: 
        data_clean_target.loc[i,'TargetClass'] = 1
     elif "B" in data_clean_target.loc[i,'TargetClass']:
         data_clean_target.loc[i,'TargetClass'] = 2
     elif "A" in data_clean_target.loc[i,'TargetClass']:
         data_clean_target.loc[i,'TargetClass'] = 3
     elif "F" in data_clean_target.loc[i,'TargetClass']:
         data_clean_target.loc[i,'TargetClass'] = 4
     elif "G" in data_clean_target.loc[i,'TargetClass']:
         data_clean_target.loc[i,'TargetClass'] = 5
     elif "K" in data_clean_target.loc[i,'TargetClass']:
         data_clean_target.loc[i,'TargetClass'] = 6
     elif "M" in data_clean_target.loc[i,'TargetClass']:
         data_clean_target.loc[i,'TargetClass'] =7
     else: 
        data_clean_target.loc[i,'TargetClass'] = 0 # None
        
data_clean_target['TargetClass']
```

```python

# Define the class labels mapping
class_labels = {
    0: 'Others',
    1: 'O',
    2: 'B',
    3:'A',
    4:'F',
    5:'G',
    6:'K',
    7:'M',
}
counts = data_clean_target['TargetClass'].value_counts()

# Map the class labels
counts.index = counts.index.map(class_labels)

# Plot the pie chart
plt.pie(counts, labels=counts.index, colors=plt.cm.Set3.colors)

# Add a title
plt.title('Distribution of samples by class')
plt.show()
```

```python
data_for_training = data_clean_target.drop(["umag","gmag","rmag","imag","zmag", "BPmag", "RPmag", "Plx"], axis=1)
data_for_training["bp-rp"] = data_for_training["r_BPmag"] - data_for_training["r_RPmag"]
```

```python
data_for_training
```

## Mapping the data on a scatter plot

```python
class_labels = ['O', 'B', 'A','F','G','K','M'] # replace with your class labels
scatter = plt.scatter(data_for_training['bp-rp'], data_for_training['r_Gmag'], c=data_for_training['TargetClass'], alpha=0.4, label = class_labels)
plt.gca().invert_yaxis()
plt.legend(handles=scatter.legend_elements()[0], labels=class_labels)
plt.xlabel ('BPmag - RPmag')
plt.ylabel ('Relative Gmag')
plt.title('Color-Magnitude Diagram (CMD) for stars')
plt.show()
```

```python
clean_target = data_for_training[data_for_training.TargetClass != 0]
file_path = "data/training_data.csv"
clean_target.to_csv(file_path, index=False)
```
